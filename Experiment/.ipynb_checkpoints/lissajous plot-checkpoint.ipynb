```python
import os
import csv
import numpy as np
import matplotlib
# matplotlib.use('Agg')  # Use the Agg backend (a non-interactive backend)
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = False


# Define the input folder path containing the CSV files
input_folder_path = "/Users/leonardo/Documents/Jupyter_Notebook_Workspace/Tracking_Data_Plotting/DataProcess/"

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(input_folder_path) if f.endswith(".csv")]

# Define the time range for cropping
start_time = 58.0*0  # Specify the start time in seconds
end_time = 4000    # Specify the end time in seconds

# Loop through each CSV file in the folder
for csv_filename in csv_files:
    csv_file_path = os.path.join(input_folder_path, csv_filename)

    def get_column_values(row, column_indices):
        return [float(value) for value in row]

    target_column_indices = [0, 1, 2, 3]  # Adjust these indices based on your header

    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = next(csv_reader)[0].split()  # Split the header into individual column names
        print("CSV Header:", header)  # Print header for debugging

        time_values, x_values, y_values, z_values, x_waypoint, y_waypoint, z_waypoint = [], [], [], [], [], [], []

        for row in csv_reader:
            values = row[0].split()  # Split the row into individual values
            data = get_column_values(values, target_column_indices)
            time = data[0]

            # Check if the time is within the specified range
            if start_time <= time <= end_time:
                time_values.append(time)
                x_values.append(data[1])
                y_values.append(data[2])
                z_values.append(data[3])
                x_waypoint.append(data[11])
                y_waypoint.append(data[12])
                z_waypoint.append(data[13])

    # Calculate errors for x, y, and z
    error_x = np.array(x_waypoint) - np.array(x_values)
    error_y = np.array(y_waypoint) - np.array(y_values)
    error_z = np.array(z_waypoint) - np.array(z_values)

    # Calculate the Euclidean error in R3
    error_r3 = np.sqrt(error_x**2 + error_y**2 + error_z**2)

    # Calculate mean and standard deviation
    mean_error_r3 = np.mean(error_r3)
    std_dev_error_r3 = np.std(error_r3)

    # Create subplots and figures
    fig1, ax1 = plt.subplots(figsize=(8, 2))
    fig2, ax2 = plt.subplots(figsize=(8, 6))

    # Plot errors in x, y, z, and R3 on the first figure
    ax1.plot(time_values, error_x, label='Error in x', color='purple')
    ax1.plot(time_values, error_y, label='Error in y', color='orange')
    ax1.plot(time_values, error_z, label='Error in z', color='green')
#     ax1.plot(time_values, error_r3, label='Error in R3', color='blue')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Error (m)')
    ax1.legend(loc='upper left')
    ax1.grid(color='k', linestyle='-', linewidth=0.2)
    ax1.set_title(f'Errors vs. Time for {csv_filename}\nMean R3 Error: {mean_error_r3:.4f} m, Std Dev R3 Error: {std_dev_error_r3:.4f} m')

    # Save the first figure to a PNG file with 200 DPI
    output_error_png_path = os.path.join(input_folder_path, f"error_figure_{csv_filename}.png")
    fig1.savefig(output_error_png_path, dpi=200, bbox_inches='tight')

    # Plot x vs y data with reference lines on the second figure
    ax2.plot(x_values, y_values, label='x vs y', color='blue')
    ax2.plot(x_waypoint, y_waypoint, label='x_waypoint vs y_waypoint', linestyle='--', color='red')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.legend(loc='upper left')
    ax2.grid(color='k', linestyle='-', linewidth=0.2)
    ax2.set_title(f'x vs y for {csv_filename}')

    # Save the second figure to a PNG file with 200 DPI
    output_xy_png_path = os.path.join(input_folder_path, f"xy_figure_{csv_filename}.png")
    fig2.savefig(output_xy_png_path, dpi=200, bbox_inches='tight')
    
    # Close the figures
    plt.close(fig1)
    plt.close(fig2)
    

# Show the plots for all CSV files
plt.show()
```

```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt

# Define the input folder path containing the CSV files
input_folder_path = "/Users/leonardo/Documents/Jupyter_Notebook_Workspace/Tracking_Data_Plotting/DataProcess/"

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(input_folder_path) if f.endswith(".csv")]

# Define the time range for cropping
start_time = 58.0 * 0  # Specify the start time in seconds
end_time = 4000  # Specify the end time in seconds

# Loop through each CSV file in the folder
for csv_filename in csv_files:
    csv_file_path = os.path.join(input_folder_path, csv_filename)

    def get_column_values(row, column_indices):
        return [float(value) for value in row]

    target_column_indices = [0, 1, 2, 3]  # Adjust these indices based on your header

    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = next(csv_reader)[0].split()  # Split the header into individual column names
        print("CSV Header:", header)  # Print header for debugging

        time_values, x_values, y_values, z_values, x_waypoint, y_waypoint, z_waypoint = [], [], [], [], [], [], []

        for row in csv_reader:
            values = row[0].split()  # Split the row into individual values
            data = get_column_values(values, target_column_indices)
            time = data[0]

            # Check if the time is within the specified range
            if start_time <= time <= end_time:
                time_values.append(time)
                x_values.append(data[1])
                y_values.append(data[2])
                z_values.append(data[3])
                x_waypoint.append(data[11])
                y_waypoint.append(data[12])
                z_waypoint.append(data[13])

    # Calculate errors for x, y, and z
    error_x = np.array(x_waypoint) - np.array(x_values)
    error_y = np.array(y_waypoint) - np.array(y_values)
    error_z = np.array(z_waypoint) - np.array(z_values)

    # Calculate the Euclidean error in R3
    error_r3 = np.sqrt(error_x ** 2 + error_y ** 2 + error_z ** 2)

    # Calculate mean and standard deviation
    mean_error_r3 = np.mean(error_r3)
    std_dev_error_r3 = np.std(error_r3)

    # Create a single figure with two subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))  # Adjust figsize for the desired aspect ratio

    # Plot errors in x, y, z, and R3 on the first subplot
    ax1.plot(time_values, error_x, label='Error in x', color='purple')
    ax1.plot(time_values, error_y, label='Error in y', color='orange')
    ax1.plot(time_values, error_z, label='Error in z', color='green')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Error (m)')
    ax1.legend(loc='upper left')
    ax1.set_aspect('20')
    ax1.grid(color='k', linestyle='-', linewidth=0.2)
    ax1.set_title(f'Errors vs. Time for {csv_filename}\nMean R3 Error: {mean_error_r3:.4f} m, Std Dev R3 Error: {std_dev_error_r3:.4f} m')

    # Plot x vs y data with reference lines on the second subplot
    ax2.plot(x_values, y_values, label='x vs y', color='blue')
    ax2.plot(x_waypoint, y_waypoint, label='x_waypoint vs y_waypoint', linestyle='--', color='red')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.legend(loc='upper left')
    ax2.set_aspect('1.25')
    ax2.grid(color='k', linestyle='-', linewidth=0.2)
    ax2.set_title(f'x vs y for {csv_filename}')

    # Adjust the subplots' spacing
    plt.tight_layout()

    # Save the figure to a PNG file with 200 DPI
    output_figure_path = os.path.join(input_folder_path, f"combined_figure_{csv_filename}.png")
    fig.savefig(output_figure_path, dpi=200)

    # Close the figure
    plt.close(fig)

# Show the plots for all CSV files
plt.show()
```
